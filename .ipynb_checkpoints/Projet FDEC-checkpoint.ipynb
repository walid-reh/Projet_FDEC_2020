```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
dfmoufia.plot(kind='scatter' , x ='FG_Avg', y = 'Text_Avg')
```

```python
# Test train split for supervised training
X_train, X_test, y_train, y_test = train_test_split(dfmoufia.FG_Avg, dfmoufia.Text_Avg)
```

```python
# Test train split visualization 
plt.scatter(X_train, y_train, label='Training Data', color = 'r', alpha=.7)
plt.scatter(X_test, y_test, label= 'Testing Data', color ='g', alpha =.7)
plt.legend()
plt.title("Test Train Split")
plt.show()
```

```python
# Creating the linear model and train it
LR= LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
# Use model to predict on TEST Data
prediction = LR.predict(X_test.values.reshape(-1,1))

#Plot prediction line against actual test data
plt.plot(X_test, prediction, label = 'Linear Regression', color ='b')
plt.scatter(X_test, y_test, label='Actual Test Data', color='g', alpha=.7)
plt.legend()
plt.show()
```

```python
# Predict a specific temperature for a FG_Avg
LR.predict(np.array([[6.017]]))[0]
```

```python
dfmoufia  = pd.read_csv(r'C:\Users\R_Walid\Desktop\Projet FDEC\Projet_FDEC_2020\data\Moufia_2014_2015\moufia_2014_2015.csv',sep=',')
dfmoufia['Timestamp']=pd.to_datetime(dfmoufia['Timestamp'])
data_hourly = dfmoufia.resample('H',on='Timestamp').mean()
```

```python
data_hourly.head()
```

```python
data_hourly["Kb"] =  data_hourly["FD_Avg"] / data_hourly["FG_Avg"]
data_hourly["FDir_Avg"] = data_hourly["FG_Avg"] - data_hourly["FD_Avg"]
```

```python
data_hourly.shape
```

```python
import matplotlib.pyplot as plt
from fbprophet.plot import plot_plotly, add_changepoints_to_plot

def plot_Flow():
    for csv in ["Moufia_2014_2015/moufia_2014_2015", "Possession_2014_2015/possession_2014_2015", "SaintAndre_2014_205/saintandre_2014_2015", "SaintLeu_2014_2015/saintleu_2014_2015", "SaintPierre_2014_2015/saintpierre_2014_2015"]:
        df = pd.read_csv('data/'+csv+'.csv', delimiter=',')
        df[['Date','Time']]=df.Timestamp.str.split(" ", expand=True)
        df['Timestamp']=pd.to_datetime(df['Timestamp'])
        df = df.resample('H',on='Timestamp').mean()
        #print(df)
        #df=df.drop(['Timestamp','Time'],axis=1)
        df["FDir_Avg"] = df["FG_Avg"] - df["FD_Avg"]
        plt.figure(figsize=(8,4), dpi=100)
        plt.plot(df['FDir_Avg'], label = csv)
        plt.legend()
        plt.title("Hourly Direct solar flow from 2014 to 2015")
        plt.xlim(pd.Timestamp('2014-01-01'), pd.Timestamp('2015-12-31'))
```
